#### 1.2 Global helper functions

### 2.4 Encoding posture, special characters, and multilingual storefront text

Here I verify how the flat file tolerates UTF-8 decoding end-to-end, sketch non-ASCII density on representative titles and blurbs, and harvest language tokens embedded in the Steam language roster column.

```python
# Section 2.4 — UTF-8 decoding, non-ASCII density, language tokens
import ast
from collections import Counter

# UTF-8 byte check
with open(DATA_PATH, "rb") as fh:
    raw_bytes = fh.read()
try:
    raw_bytes.decode("utf-8")
    utf8_strict_ok = True
except UnicodeDecodeError as exc:
    utf8_strict_ok = False
    print("【test log】 UTF-8 strict decode error:", exc)
print("【test log】 entire file UTF-8 strict decode OK:", utf8_strict_ok)


# non-ASCII share helper
def non_ascii_ratio(series: pd.Series) -> float:
    s = series.dropna().astype(str)
    if s.empty:
        return float("nan")
    mask = s.str.contains(r"[^\x00-\x7F]", regex=True)
    return float(mask.mean())


# storefront text scan
for col in ["Name", "About the game", "Supported languages"]:
    if col in df.columns:
        r = non_ascii_ratio(df[col])
        print(f"【test log】 non-ASCII row fraction — {col}:", round(r, 4))


# parse language list cell
def parse_language_cell(val) -> list:
    if pd.isna(val):
        return []
    text = str(val).strip()
    if text in ("", "[]"):
        return []
    try:
        obj = ast.literal_eval(text)
        if isinstance(obj, list):
            return [str(x).strip() for x in obj if str(x).strip()]
    except (ValueError, SyntaxError):
        pass
    return []


# top language tokens
lang_counter = Counter()
for cell in df["Supported languages"].dropna():
    for token in parse_language_cell(cell):
        lang_counter[token] += 1
lang_freq = (
    pd.DataFrame(lang_counter.most_common(25), columns=["language_token", "listing_rows"])
)
display(lang_freq)
```

Reading the file as raw bytes succeeds under strict UTF-8 end-to-end, matching the encoding argument already supplied to `read_csv`, so there is no evidence here of hidden Latin-1 fallback during ingestion. Non-Latin titles or blurbs appear on a measurable minority of `Name` rows and more frequently inside long descriptions, which is expected for a global storefront dump; the ranked frequency table surfaces storefront tokens such as English, Simplified Chinese, German, French, and Japanese exactly as Steam exports them, rather than inferring locale from glyphs alone.

```python
# Section 2.5 — nulls, duplicates, replacement chars, ordering, dtype mismatches
from IPython.display import display

# missingness table
miss_tbl = missingness_summary(df)
display(miss_tbl)

# duplicate key check
dup_appid = int(df["AppID"].duplicated().sum())
dup_name = int(df["Name"].duplicated().sum())
print("【test log】 duplicate AppID rows:", dup_appid)
print("【test log】 duplicate Name rows:", dup_name)

# U+FFFD replacement scan
replacement_rows = {}
for col in df.select_dtypes(include=["object", "string"]).columns:
    ser = df[col].dropna().astype(str)
    if ser.empty:
        continue
    cnt = int(ser.str.contains("\ufffd", regex=False).sum())
    if cnt:
        replacement_rows[col] = cnt
print("【test log】 U+FFFD replacement counts by column:", replacement_rows)

print("【test log】 AppID strictly increasing along row order?", bool(df["AppID"].is_monotonic_increasing))

# numeric-coercible strings
parseable_share = []
for col in df.select_dtypes(include=["object", "string"]).columns:
    ser = df[col].dropna()
    if ser.empty:
        continue
    coerced = pd.to_numeric(ser, errors="coerce")
    frac = float(coerced.notna().mean())
    if frac >= 0.95:
        parseable_share.append((col, round(frac, 4)))
parseable_share.sort(key=lambda t: -t[1])
print("【test log】 object/string columns with >=95% numeric-parseable values:")
for col, frac in parseable_share[:10]:
    print(col, frac)
if not parseable_share:
    print("【test log】 no object/string column reaches >=95% numeric coercion")
```

Missingness concentrates in storefront-media columns (`Movies`, `Score rank`, review bodies) rather than core telemetry, while `AppID` repeats never occur yet roughly **1,156** display titles collide—consistent with regional duplicates rather than a broken surrogate key. Replacement-character tallies stay negligible outside isolated long-text cells, so wholesale mojibake is not indicated at this grain. Row order is not sorted by `AppID`, meaning sequential scans treat the table as append-only catalogue order rather than ascending identifier sequence. Finally, almost every object column fails universal coercion to numeric except where Steam genuinely stores scalars as strings; nothing in this pass suggests pandas attached the wrong elemental dtype class beyond those already inspected.

#### 2.6a — build mining frame and correlation heatmap

Before plotting, a Pearson correlation heatmap on the §2.6 core numeric block tests which engagement, commerce, and playtime fields move together catalogue-wide. The matrix uses parsed owner midpoints and vote-derived `review_ratio` alongside raw counters.

```python
# Section 2.6a — build mining frame and correlation heatmap
from scipy import stats


# owner range midpoint
def owner_band_midpoint(label):
    """Map Steam owner range text to the midpoint of the bracket."""
    if pd.isna(label):
        return np.nan
    text = str(label).strip()
    if "-" not in text:
        return np.nan
    lo_s, hi_s = text.split("-", maxsplit=1)
    try:
        lo = float(lo_s.strip().replace(",", ""))
        hi = float(hi_s.strip().replace(",", ""))
        return (lo + hi) / 2.0
    except ValueError:
        return np.nan


# mining frame
mine = pd.DataFrame(index=df.index)
mine["owner_mid"] = df["Estimated owners"].map(owner_band_midpoint)
for col in [
    "Price",
    "Discount",
    "Peak CCU",
    "Positive",
    "Negative",
    "Recommendations",
    "Average playtime forever",
    "Median playtime forever",
    "Metacritic score",
    "Achievements",
    "DLC count",
]:
    mine[col] = df[col]

# player vote ratio
vote_total = mine["Positive"] + mine["Negative"]
mine["review_ratio"] = np.where(vote_total > 0, mine["Positive"] / vote_total, np.nan)

corr_cols = [
    "Price",
    "Discount",
    "Peak CCU",
    "Positive",
    "review_ratio",
    "Metacritic score",
    "owner_mid",
    "Average playtime forever",
    "Median playtime forever",
    "Recommendations",
    "Achievements",
    "DLC count",
]
# Pearson correlation
corr_mat = mine[corr_cols].corr(method="pearson")
print("【test log】 correlation matrix shape:", corr_mat.shape)

fig, ax = plt.subplots(figsize=(10, 8))
# correlation heatmap
sns.heatmap(corr_mat, annot=True, fmt=".2f", cmap="coolwarm", center=0, ax=ax)
ax.set_title("Pearson correlation — core numeric features")
rotate_x_labels_if_needed(ax)
plt.tight_layout()
plt.show()
```

The heatmap shows **`Positive`** and **`Peak CCU`** sharing a strong positive cell (**0.85**), and **`Recommendations`** with **`Positive`** at **0.90**. **`Average playtime forever`** and **`Median playtime forever`** are almost collinear (**0.99**), while **`Discount`** against **`Price`** is only **−0.05**—essentially flat at catalogue scale.

#### 2.6b — log-scaled engagement scatter

A log-scaled scatter of **`Positive`** against **`Peak CCU`** checks whether the heatmap link survives compression of seven-digit tails noted in §2.3.

#### 2.6e NHST — do primary genres differ in list price?

§2.6d already shows different **median `Price`** bars by primary genre token. This step runs a **null-hypothesis significance test (NHST)** on a narrower question only: are those bar gaps plausibly due to sampling noise alone? **Kruskal–Wallis** compares raw **`Price`** across the same eight primary genres (each **≥ 500** listings, **Free To Play** excluded) without assuming normality; at α = **0.05**, H₀ is that all genre **`Price`** distributions are identical and H₁ is that at least one group differs in rank/location. A **Mann–Whitney U** follow-up then contrasts **Simulation/RPG** with **Action/Adventure/Indie/Casual** because §2.6d visually split those tiers (~**$4.99** vs ~**$1.99–$2.99**). This step does **not** claim that genre *causes* price, that genre predicts price in a model, or that genre adds signal *independently of* engagement fields already correlated in §2.6a–§2.6c—it only asks whether cross-genre price ranks differ enough to reject a common-distribution null.

```python
# Section 2.6e — NHST: Price by primary genre (Kruskal–Wallis + Mann–Whitney)
from scipy import stats

ALPHA = 0.05


def primary_genre_token_nhst(cell):
    """First comma-separated genre token (same rule as §2.6d)."""
    if pd.isna(cell):
        return np.nan
    return str(cell).split(",")[0].strip()


nhst_genre = df.copy()
nhst_genre["genre_primary"] = nhst_genre["Genres"].map(primary_genre_token_nhst)

# eligible genre groups
genre_counts = nhst_genre.groupby("genre_primary").size()
eligible_genres = genre_counts[genre_counts >= 500].index.difference(["Free To Play"])

genre_samples = []
for genre in sorted(eligible_genres):
    prices = nhst_genre.loc[nhst_genre["genre_primary"] == genre, "Price"].dropna()
    genre_samples.append(prices.values)

kw_stat, kw_p = stats.kruskal(*genre_samples)
print("【test log】 Kruskal–Wallis groups (n>=500, excl. Free To Play):", list(sorted(eligible_genres)))
print(
    "【test log】 Kruskal–Wallis on Price:",
    f"H={kw_stat:.4f}, p={kw_p:.4e}, alpha={ALPHA}",
)

# focused pairwise contrast (§2.6d high-tier vs mass-tier)
high_tier_mask = nhst_genre["genre_primary"].isin(["Simulation", "RPG"])
mass_tier_mask = nhst_genre["genre_primary"].isin(
    ["Action", "Adventure", "Indie", "Casual"]
)
price_high = nhst_genre.loc[high_tier_mask, "Price"].dropna()
price_mass = nhst_genre.loc[mass_tier_mask, "Price"].dropna()
mw_stat, mw_p = stats.mannwhitneyu(
    price_high, price_mass, alternative="two-sided"
)
print(
    "【test log】 Mann–Whitney Simulation/RPG vs Action/Adventure/Indie/Casual:",
    f"n_high={len(price_high)}, n_mass={len(price_mass)}, p={mw_p:.4e}",
)
```

**Kruskal–Wallis** across **eight** primary genres (each **≥ 500** titles, **Free To Play** excluded) returns **H = 1906.95** and **p ≈ 0** at α = **0.05**, so H₀ is rejected: the median bar gaps in §2.6d are not plausibly chart noise alone. **Mann–Whitney** on **Simulation/RPG** (**n = 4,359**) versus **Action/Adventure/Indie/Casual** (**n = 104,490**) also rejects H₀ (**p = 1.44×10⁻¹³⁸**), matching the **test log** medians (~**$4.99** vs ~**$1.99–$2.99**). With catalogue-scale sample sizes, extremely small p-values are expected even when the practical spread is modest; the test corroborates the descriptive split in §2.6d but does **not** prove genre is an independent commercial driver separate from confounded engagement or pricing norms. Retaining **`Genres`** (or a primary token) remains sensible for descriptive stratification in preprocessing, while causal or incremental predictive claims belong in supervised modelling later. §2.7 summarises §2.6; Part 3 drops sparse media fields and keeps parsed owner/playtime features before modelling.

Section 2.6 therefore points preprocessing toward dropping empty media fields, parsing owner bands and release dates, keeping one playtime aggregate, and deferring sparse **`User score`** until an imputation rule exists. **§2.6e** adds a rank-based check that primary-genre medians in §2.6d are not mere visual noise; it does **not** by itself establish genre as a standalone predictor orthogonal to the engagement block in §2.6a–§2.6c. Part 4 then tests which *prior* commercial fields predict **`is_active`**.

```python
# Section 3.2 — defensive missing handling; create df_foundation
import re
import numpy as np

# mutable copy
work = df_prep.copy()
rows_before = len(work)

# Step 1: drop hollow catalogue rows (narrow rule; reduces row count before parsing)
TAXONOMY_COLS = ["Genres", "Categories", "Developers", "Publishers", "Tags"]
# hollow-row flag
all_taxonomy_na = work[TAXONOMY_COLS].isna().all(axis=1)
has_engagement = (
    (work["Positive"] + work["Negative"] > 0)
    | (work["Peak CCU"] > 0)
    | (work["Average playtime forever"] > 0)
)
bad_row = work["Name"].isna() | (all_taxonomy_na & ~has_engagement)

print("【test log】 rows before row prune:", rows_before)
print("【test log】 bad rows (Name NA or hollow taxonomy + no engagement):", int(bad_row.sum()))
print("【test log】 bad rows with all taxonomy NA:", int((bad_row & all_taxonomy_na).sum()))
print("【test log】 bad rows with Name NA:", int(work.loc[bad_row, "Name"].isna().sum()))

# drop hollow rows
work = work.loc[~bad_row].copy()
rows_after = len(work)

# Step 2: parse Estimated owners (keep original interval column)
# owner regex
OWNER_PATTERN = re.compile(r"([\d,]+)\s*-\s*([\d,]+)")


def parse_owner_bounds(text):
    if pd.isna(text):
        return np.nan, np.nan
    match = OWNER_PATTERN.search(str(text))
    if not match:
        return np.nan, np.nan
    lower = int(match.group(1).replace(",", ""))
    upper = int(match.group(2).replace(",", ""))
    return lower, (lower + upper) / 2.0


# parse owner bounds
parsed = work["Estimated owners"].map(parse_owner_bounds)
work["owner_lowerbound"] = parsed.map(lambda x: x[0])
work["owner_midpoint"] = parsed.map(lambda x: x[1])

owner_nonempty = work["Estimated owners"].notna()
parse_fail = owner_nonempty & work["owner_lowerbound"].isna()
print("【test log】 Estimated owners non-null:", int(owner_nonempty.sum()))
print("【test log】 owner regex parse failures:", int(parse_fail.sum()))

# Step 3: missing indicators for columns with NA rate >= 5% on df_prep (no fill on originals)
# has_* indicators
HIGH_NA_COLS = [
    "Tags",
    "Categories",
    "Publishers",
    "About the game",
    "Developers",
    "Genres",
]
HAS_COLS = []
for col in HIGH_NA_COLS:
    flag = "has_" + col.lower().replace(" ", "_")
    work[flag] = work[col].notna().astype(int)
    HAS_COLS.append(flag)

print("【test log】 has_* columns added:", HAS_COLS)

# Guard: continuous telemetry must stay unfilled
# no fill on telemetry
FORBID_FILL = [
    "Price",
    "Positive",
    "Negative",
    "Peak CCU",
    "Average playtime forever",
    "Average playtime two weeks",
    "Recommendations",
    "Discount",
]
na_on_continuous = work[FORBID_FILL].isna().sum()
print("【test log】 NA on continuous columns (expect all 0):", na_on_continuous.to_dict())

# checkpoint save
df_foundation = dataframe_checkpoint(work)
print("【test log】 rows after prune:", rows_after)
print("【test log】 df_foundation.shape:", df_foundation.shape)
print("【test log】 new columns vs df_prep:", df_foundation.shape[1] - df_prep.shape[1])
```

Row pruning removes **7,982** hollow listings (**6.51%** of `df_prep`), leaving **114,629** rows. **`df_foundation`** adds **8** columns: **`owner_lowerbound`**, **`owner_midpoint`**, and six **`has_*`** indicators; shape is **(114,629, 36)**. Original taxonomy text and interval **`Estimated owners`** remain; continuous telemetry is untouched. This checkpoint is the shared entry for later supervised, unsupervised, or time-series branches.

The code below re-quantifies skew, zero share, and Tukey counts on **`df_foundation`** (**114,629** rows).

```python
# Section 3.3 — outlier recap audit on df_foundation
from scipy import stats

OUTLIER_AUDIT_COLS = [
    "Average playtime forever",
    "Average playtime two weeks",
    "Peak CCU",
    "Positive",
    "Negative",
    "Recommendations",
    "Price",
    "Discount",
    "Metacritic score",
    "User score",
    "Achievements",
    "DLC count",
    "owner_lowerbound",
    "owner_midpoint",
]

# IQR / z-score loop
audit_rows = []
for col in OUTLIER_AUDIT_COLS:
    ser = df_foundation[col]
    q1, q3 = ser.quantile(0.25), ser.quantile(0.75)
# Tukey fences
    iqr = q3 - q1
    upper = q3 + 1.5 * iqr
    lower = q1 - 1.5 * iqr
    z = np.abs(stats.zscore(ser.astype(float)))
    audit_rows.append(
        {
            "column": col,
            "zero_pct": round(float((ser == 0).mean() * 100), 2),
            "median": float(ser.median()),
            "p99": float(ser.quantile(0.99)),
            "max": float(ser.max()),
            "skew": round(float(stats.skew(ser.astype(float))), 2),
            "tukey_high": int((ser > upper).sum()),
            "tukey_low": int((ser < lower).sum()),
            "z_gt_3": int((z > 3).sum()),
        }
    )

# sorted audit table
audit_df = pd.DataFrame(audit_rows).sort_values("skew", ascending=False)
print("【test log】 df_foundation rows:", df_foundation.shape[0])
print("【test log】 outlier audit (sorted by skew):")
print(audit_df.to_string(index=False))

pos_mega = int((df_foundation["Positive"] > 1_000_000).sum())
price_disc = int(
    ((df_foundation["Price"] == 0) & (df_foundation["Discount"] > 0)).sum()
)
print("【test log】 Positive > 1,000,000:", pos_mega)
print("【test log】 Price == 0 and Discount > 0:", price_disc)
```

Tukey-high counts are large (**~19k–39k** rows) because zero inflation collapses the IQR — they mark **tail mass**, not a small bad cluster. **`Discount`** has **0** Tukey-high hits here; **39** rows with `Price = 0` and `Discount > 0` are **logic** outliers to flag separately. ## Part 4: Decision-tree attribution of game activity drivers

**`df_foundation`** is the modelling entry (**114,576** rows after Part 3). ### 4.1 Target variable and prior feature matrix

**Target \(y\) — catalogue activity (binary).** **`Reviews`** was dropped in §3.1. **`is_active = 1`** when **`Positive + Negative > 0`** or **`Peak CCU > 0`**.

**Features \(X\) — prior listing attributes only.** \(X\) keeps eight fields fixed **before** player engagement unfolds: **`Price`**, **`Discount`**, **`DLC count`**, **`Achievements`**, **`Required age`**, **`Windows`**, **`Mac`**, **`Linux`**. #### Data leakage audit — why `has_tags` must leave \(X\)

A first pass included §3.2 **`has_*`** flags (e.g. **`has_tags`**) alongside commercial fields. **`has_tags`** records whether the scraper observed a tag string **before** §3.4 imputation — a **post-catalogue / scrape-state** variable, not a design choice a developer makes to “cause” heat. It is **near-deterministic** for \(y\): listings with **`has_tags = 0`** are **~97%** inactive, while tagged rows are **~99.6%** active. The tree then split on **`has_tags`** first (Gini **≈ 0.995**), test **accuracy ≈ 0.987** — a **tautology**, not pricing insight.

**Fix:** drop **`has_tags`**, **`has_categories`**, **`has_publishers`**, **`has_about_the_game`**, **`has_developers`**, **`has_genres`**, and any similar “field was non-null” indicators from \(X\). Re-fit on **prior features only**; expect **lower accuracy** but **interpretable** splits on price, discount, and achievement depth.

```python
# Section 4.1 — is_active, prior-only X (post leakage audit), stratified split
from sklearn.model_selection import train_test_split
from sklearn.tree import DecisionTreeClassifier, plot_tree
from sklearn.metrics import accuracy_score, classification_report

EXCLUDED_METADATA_FLAGS = [
    "has_tags",
    "has_categories",
    "has_publishers",
    "has_about_the_game",
    "has_developers",
    "has_genres",
    "has_metacritic",
]

FEATURE_COLS = [
    "Price",
    "Discount",
    "DLC count",
    "Achievements",
    "Required age",
    "Windows",
    "Mac",
    "Linux",
]

LEAKAGE_COLS = [
    "Positive",
    "Negative",
    "Recommendations",
    "Peak CCU",
    "Average playtime forever",
    "Average playtime two weeks",
    "Metacritic score",
    "User score",
    "owner_midpoint",
    "owner_lowerbound",
] + EXCLUDED_METADATA_FLAGS

# modelling frame
df_model = df_foundation.copy()
# binary target
df_model["is_active"] = (
    (df_model["Positive"] + df_model["Negative"] > 0) | (df_model["Peak CCU"] > 0)
).astype(int)

y = df_model["is_active"]
print("【test log】 modelling rows:", len(df_model))
print("【test log】 is_active rate:", round(float(y.mean()), 4))
print("【test log】 inactive / active:", int((y == 0).sum()), int((y == 1).sum()))

for col in FEATURE_COLS:
    if col not in df_model.columns:
        raise KeyError(f"Missing prior feature: {col}")

# leakage guard
overlap = [c for c in LEAKAGE_COLS if c in FEATURE_COLS]
if overlap:
    raise ValueError(f"Leakage columns still in FEATURE_COLS: {overlap}")

print("【test log】 excluded metadata flags:", EXCLUDED_METADATA_FLAGS)
print("【test log】 prior feature count:", len(FEATURE_COLS))
print("【test log】 prior feature names:", FEATURE_COLS)

# prior feature matrix
X = df_model[FEATURE_COLS].fillna(0)

# stratified split
X_train, X_test, y_train, y_test = train_test_split(
    X,
    y,
    test_size=0.25,
    random_state=RANDOM_STATE,
    stratify=y,
)
print("【test log】 train shape:", X_train.shape, "test shape:", X_test.shape)
print(
    "【test log】 train is_active rate:",
    round(float(y_train.mean()), 4),
    "test:",
    round(float(y_test.mean()), 4),
)
```

On **`df_foundation`**, **`is_active`** covers **~73.4%** of rows (**84,045** active vs **30,531** inactive). After the audit, \(X\) has **eight** prior fields only — no **`has_*`** scrape flags, no owner tiers, no review scores. The stratified hold-out keeps the same class mix for a fair re-fit.

### 4.2 Decision tree classifier (leakage-free re-fit)

The same **`DecisionTreeClassifier`** settings (**`max_depth=5`**, **`min_samples_leaf=200`**, **`RANDOM_STATE`**) are applied to the **prior-only** matrix. Accuracy should **fall** versus the tautological model — that drop is expected and desirable. The **classification report** on the test fold shows whether inactive listings (minority) remain detectable from commercial attributes alone.

```python
# Section 4.2 — fit decision tree on prior features only
# shallow tree
dt_model = DecisionTreeClassifier(
    max_depth=5,
    min_samples_leaf=200,
    random_state=RANDOM_STATE,
)
# train on priors
dt_model.fit(X_train, y_train)

# hold-out predict
y_pred = dt_model.predict(X_test)
test_acc = accuracy_score(y_test, y_pred)

print("【test log】 tree max_depth param:", dt_model.max_depth)
print("【test log】 fitted tree depth:", dt_model.get_depth())
print("【test log】 leaf count:", dt_model.get_n_leaves())
print("【test log】 test accuracy:", round(test_acc, 4))
print("【test log】 classification report (test):")
print(classification_report(y_test, y_pred, target_names=["inactive", "active"]))
```

### 4.6 Micro-rules and misclassification audit

Section 4.5 interpreted global importances. This step extracts **explicit IF–THEN paths** to the highest-purity leaves of the depth-capped tree, then builds an **error notebook** on the hold-out fold: **false positives** (predicted active, actually inactive) and **false negatives** (predicted inactive, actually active). Tag-token frequencies on those errors hint at genres the price/discount prior misses.

#### 4.6a Explicit leaf rules

The code walks the fitted **`dt_model`** and prints readable rules for leaves with at least **`MIN_LEAF_SAMPLES`** training rows (aligned with **`min_samples_leaf=200`**), ranked by **purity** (majority-class share). Paths show how **Price**, **Discount**, and **Achievements** combine before a leaf calls **Active** or **Inactive**.

```python
# Section 4.6a — extract high-purity leaf rules as text
from sklearn.tree import _tree

MIN_LEAF_SAMPLES = 200
TOP_RULES = 12

# sklearn tree struct
tree = dt_model.tree_
feat_names = FEATURE_COLS


def _fmt_threshold(name, op, thr):
    if name in ("Price", "Discount"):
        return f"{name} {op} {thr:.2f}"
    if name in ("Achievements", "DLC count"):
        return f"{name} {op} {thr:.1f}"
    return f"{name} {op} {int(round(thr))}"


# DFS leaf walker
def _collect_leaf_rules(node_id, conditions, out):
    if tree.children_left[node_id] == tree.children_right[node_id]:
        n_samples = int(tree.n_node_samples[node_id])
        if n_samples < MIN_LEAF_SAMPLES:
            return
        counts = tree.value[node_id].reshape(-1)
        purity = float(counts.max() / counts.sum())
        label = "Active" if counts[1] >= counts[0] else "Inactive"
        out.append(
            {
                "conditions": conditions.copy(),
                "label": label,
                "purity": purity,
                "n": n_samples,
                "active_share": float(counts[1] / counts.sum()),
            }
        )
        return
    feat = feat_names[tree.feature[node_id]]
    thr = tree.threshold[node_id]
    left_id = tree.children_left[node_id]
    right_id = tree.children_right[node_id]
    _collect_leaf_rules(left_id, conditions + [(feat, "<=", thr)], out)
    _collect_leaf_rules(right_id, conditions + [(feat, ">", thr)], out)


# collect + sort rules
leaf_rules = []
_collect_leaf_rules(0, [], leaf_rules)
leaf_rules.sort(key=lambda r: (r["purity"], r["n"]), reverse=True)
leaf_rules_df = pd.DataFrame(leaf_rules)

print("【test log】 total leaves (n >= %d):" % MIN_LEAF_SAMPLES, len(leaf_rules))
print("【test log】 top leaf rules by purity:\n")
for i, rule in enumerate(leaf_rules[:TOP_RULES], start=1):
    cond_text = " and ".join(
        _fmt_threshold(name, op, thr) for name, op, thr in rule["conditions"]
    )
    print(
        f"  {i}. If {cond_text} -> {rule['label']} "
        f"(purity={rule['purity']:.3f}, n={rule['n']}, active_share={rule['active_share']:.3f})"
    )
```

```python
# Section 4.6a (figure) — top leaf purity bar chart (supplement to rule text above)
from matplotlib.patches import Patch

# top leaves to plot
TOP_PLOT = 12
# sort for barh
plot_rules = (
    leaf_rules_df.sort_values(["purity", "n"], ascending=False)
    .head(TOP_PLOT)
    .iloc[::-1]
)
palette = {"Active": "steelblue", "Inactive": "indianred"}

fig, ax = plt.subplots(figsize=(9, 5.5))
# purity bars
ax.barh(
    range(len(plot_rules)),
    plot_rules["purity"],
    color=plot_rules["label"].map(palette),
    edgecolor="0.25",
)
ax.set_yticks(range(len(plot_rules)))
ax.set_yticklabels(
    [f"{lbl} | n={n:,}" for lbl, n in zip(plot_rules["label"], plot_rules["n"])],
    fontsize=9,
)
ax.set_xlim(0, 1.05)
ax.set_xlabel("Leaf purity (majority-class share)")
ax.set_title("Decision tree — top 12 training leaves by purity (n ≥ 200)")
for i, pur in enumerate(plot_rules["purity"]):
    ax.text(pur + 0.008, i, f"{pur:.3f}", va="center", fontsize=8)
ax.legend(
    handles=[Patch(facecolor=palette[k], label=k) for k in palette],
    loc="lower right",
)
plt.tight_layout()
plt.show()
```

Among **29** training leaves with **n ≥ 200**, the **top 12 by purity are all Active**. The strongest pocket is **If `Price > 0.25` and `Discount > 51.5%` and `Achievements > 20.5` and `Price > 0.59` → Active** (**purity = 0.972**, **n = 5,000**). Nearby sale paths show **n = 3,921** at **0.927** and **n = 4,911** at **0.870** purity — paid SKUs on deep discounts. The largest high-purity active leaf in the top list is **n = 15,555** (**purity = 0.750**: moderate **`Price`**, **`Discount ≤ 40.5%`**, **`Mac = 0`**). The dominant **Inactive** leaf is **`Price ≤ 0.25`**, low **`Achievements`**, no **`Mac`**, **`DLC count ≤ 0.5`** (**n = 7,228**, **purity = 0.579**, **active share = 0.421**) — the free/low-scaffold long tail, not the purity-ranked head.

#### 4.6b Error notebook (test fold)

Misclassified hold-out rows are flagged. **False positives** are ranked by **`p_active`** (model most confident); **false negatives** by lowest **`p_active`**. For each group the code prints **`Name`**, **`Price`**, **`Tags`** (top 10) and tokenises comma-separated tags for a quick frequency table.

```python
# Section 4.6b — false positives / false negatives + tag token stats
from collections import Counter

# Rebuild the same test indices (identical to §4.1 split)
# same split indices
_, _, _, _, idx_train, idx_test = train_test_split(
    X,
    y,
    df_model.index,
    test_size=0.25,
    random_state=RANDOM_STATE,
    stratify=y,
)

# hold-out audit frame
audit = df_model.loc[idx_test].copy()
audit["y_true"] = y_test.values
audit["y_pred"] = dt_model.predict(X_test)
audit["p_active"] = dt_model.predict_proba(X_test)[:, 1]

# misclassified rows
mis = audit.loc[audit["y_true"] != audit["y_pred"]].copy()
# false positives
fp = mis.loc[(mis["y_pred"] == 1) & (mis["y_true"] == 0)].sort_values(
    "p_active", ascending=False
)
# false negatives
fn = mis.loc[(mis["y_pred"] == 0) & (mis["y_true"] == 1)].sort_values(
    "p_active", ascending=True
)

n_correct = int((audit["y_true"] == audit["y_pred"]).sum())

print("【test log】 test rows:", len(audit))
print("【test log】 misclassified:", len(mis), "FP:", len(fp), "FN:", len(fn))

cols_show = ["Name", "Price", "Tags", "p_active"]
print("\n【test log】 False positives (predicted Active, actual Inactive) — top 10:")
print(fp[cols_show].head(10).to_string(index=False, max_colwidth=50))

print("\n【test log】 False negatives (predicted Inactive, actual Active) — top 10:")
print(fn[cols_show].head(10).to_string(index=False, max_colwidth=50))


# split tag string
def _tag_tokens(tag_str):
    if pd.isna(tag_str) or str(tag_str).strip() in ("", "Unknown"):
        return []
    return [t.strip().lower() for t in str(tag_str).split(",") if t.strip()]


def _top_tag_counts(frame, top_k=10):
    counter = Counter()
    for tag_str in frame["Tags"]:
        counter.update(_tag_tokens(tag_str))
    return counter.most_common(top_k)


fp_top = _top_tag_counts(fp)
fn_top = _top_tag_counts(fn)
fp_all_top = _top_tag_counts(mis.loc[(mis["y_pred"] == 1) & (mis["y_true"] == 0)])

print("\n【test log】 FP tag tokens (top 10 among FP misclassifications):")
print(fp_all_top)
print("【test log】 FN tag tokens (top 10 among FN misclassifications):")
print(fn_top)

# For optional figure cells below
# for figure cells
fp_tags_df = pd.DataFrame(fp_all_top, columns=["tag", "count"])
fn_tags_df = pd.DataFrame(fn_top, columns=["tag", "count"])
```

```python
# Section 4.6b (figure 1) — hold-out outcome counts (supplement to tables above)
# correct / FP / FN counts
outcome_counts = pd.Series(
    {
        "Correct": n_correct,
        "False Positive": len(fp),
        "False Negative": len(fn),
    }
)
outcome_colors = {
    "Correct": "#2ecc71",
    "False Positive": "#e74c3c",
    "False Negative": "#f39c12",
}

fig, ax = plt.subplots(figsize=(6.5, 4))
# outcome bar chart
bars = ax.bar(
    outcome_counts.index,
    outcome_counts.values,
    color=[outcome_colors[k] for k in outcome_counts.index],
    edgecolor="0.2",
)
ax.set_ylabel("Test listings (n)")
ax.set_title("§4.4 hold-out — prediction outcomes")
for bar, val in zip(bars, outcome_counts.values):
    ax.text(
        bar.get_x() + bar.get_width() / 2,
        bar.get_height() + 150,
        f"{val:,}",
        ha="center",
        va="bottom",
        fontsize=10,
    )
rotate_x_labels_if_needed(ax)
plt.tight_layout()
plt.show()
```

```python
# Section 4.6b (figure 2) — top tag tokens among FP vs FN (supplement to counts above)
# FP vs FN tag bars
fig, axes = plt.subplots(1, 2, figsize=(11, 4.5))

# side-by-side panels
for ax, df_tags, title, color in [
    (axes[0], fp_tags_df, "False positives — tag tokens", "#e74c3c"),
    (axes[1], fn_tags_df, "False negatives — tag tokens", "#f39c12"),
]:
    if df_tags.empty:
        ax.text(0.5, 0.5, "No parseable tags", ha="center", va="center")
        ax.set_axis_off()
        continue
    plot_t = df_tags.sort_values("count", ascending=True)
    ax.barh(plot_t["tag"], plot_t["count"], color=color, edgecolor="0.2", alpha=0.85)
    ax.set_xlabel("Token count")
    ax.set_title(title)
    ax.set_ylabel("Tag token")

plt.suptitle("Misclassified hold-out — Tags word frequency (top 10)", y=1.02)
plt.tight_layout()
plt.show()
```

On **28,644** test rows, **7,296** are misclassified (**25.5%**): **5,962** false positives vs **1,334** false negatives. ### 4.7 Chapter summary and next steps

In sum, the depth-capped decision tree on **prior numeric features** reaches useful but asymmetric test performance (**accuracy ≈ 0.75**), yet two error regimes expose its limits. **False positives** are “parameter-gaming reskins”: listings with commercial **`Price`** / discount signals that route into high-purity **Active** leaves while **`Tags`** are often **`Unknown`** and votes/CCU stay at zero. **False negatives** are “free indies that break out via theme/IP”: **`Price = 0`** titles that remain active through **VR**, **free to play**, **early access**, and other tag-rich niches the price prior cannot see.

The residual audit’s tag-token counts make the gap explicit — **singleplayer / casual / indie** on FPs versus **indie / free to play / action / VR** on FNs — showing that **text labels**, not list price alone, explain the misclassified tail. Numeric-only modelling therefore cannot fully unpack Steam’s catalogue.

## Part 5: Text mining and unsupervised clustering

§4.6 showed that **false positives** and **false negatives** are driven by **tag semantics**, not list price alone. This chapter vectorises **`Tags`** with **TF-IDF**, selects a cluster count via the **WCSS elbow**, fits **K-Means**, and profiles each **`Cluster_ID`** against price/discount/activity — then cross-audits §4.4 FP/FN indices.

### 5.1 Text feature extraction and TF-IDF matrix

**`Tags`** on **`df_foundation`** are cleaned (`NaN` / **`Unknown`** → empty string), then passed to **`TfidfVectorizer`** (`max_features=100`, English stop words, hyphen-friendly token pattern). The sparse matrix is densified into **`df_tfidf`** with the **same index** as **`df_foundation`** for downstream joins.

```python
# Section 5.1 — TF-IDF on Tags (aligned to df_foundation)
from sklearn.feature_extraction.text import TfidfVectorizer

# clean Tags text
tags_clean = (
    df_foundation["Tags"]
    .replace("Unknown", "")
    .fillna("")
    .astype(str)
    .str.strip()
)

# non-empty tag rows
nonempty_mask = tags_clean.str.len() > 0
print("【test log】 df_foundation rows:", len(df_foundation))
print("【test log】 tags empty-string rows:", int((~nonempty_mask).sum()))
print("【test log】 tags non-empty rows:", int(nonempty_mask.sum()))

if not nonempty_mask.any():
    raise ValueError(
        "All Tags are empty after cleaning — run Part 3 through §3.4 on df_foundation first."
    )

# 100-term vocabulary
tfidf_vectorizer = TfidfVectorizer(
    max_features=100,
    stop_words="english",
    token_pattern=r"(?u)\b[a-zA-Z0-9_-]+\b",
)

# Fit vocabulary on rows with real tag text; transform full catalogue (empty -> zero vector)
# fit on non-empty only
tfidf_vectorizer.fit(tags_clean.loc[nonempty_mask])
X_tfidf_sparse = tfidf_vectorizer.transform(tags_clean)
feature_names_tfidf = tfidf_vectorizer.get_feature_names_out()

# dense TF-IDF table
df_tfidf = pd.DataFrame(
    X_tfidf_sparse.toarray(),
    index=df_foundation.index,
    columns=feature_names_tfidf,
)

print("【test log】 df_tfidf shape:", df_tfidf.shape)
print(
    "【test log】 index aligned:",
    bool(df_tfidf.index.equals(df_foundation.index)),
)
print("【test log】 first 15 TF-IDF terms:", list(feature_names_tfidf[:15]))
```

After vectorisation, **`df_tfidf`** is **(114,576 × 100)** with index parity to **`df_foundation`**. A large share of rows have **empty** tag strings post-cleaning (§3.4 **`Unknown`** placeholders), so many vectors are sparse zeros — those listings will be pulled toward a “tag-poor” centroid later. The first extracted terms include genre staples (**`2d`**, **`3d`**, **`action`**, **`adventure`**, **`casual`**, …), confirming the vocabulary captures Steam storefront phrasing.

### 5.2 Unsupervised tuning: WCSS elbow curve

**`KMeans`** is fit for **K = 2…8** on **`X_tfidf_sparse`** (same representation as **`df_tfidf`**). **`inertia_`** (WCSS) is plotted vs **K**; the bend near **K = 5** is used for the final clustering pass (`OPT_K = 5` unless manually overridden after inspecting the curve).

```python
# Section 5.2 — WCSS elbow for K in 2..8 (single figure)
# elbow sweep
K_RANGE = range(2, 9)
wcss_by_k = []

# WCSS per K
for k in K_RANGE:
    km_probe = KMeans(n_clusters=k, random_state=RANDOM_STATE, n_init=10)
    km_probe.fit(X_tfidf_sparse)
    wcss_by_k.append(km_probe.inertia_)
    print(f"【test log】 K={k} WCSS (inertia)={km_probe.inertia_:.2f}")

# Default operating K after visual elbow (override if the plot suggests otherwise)
# operating K
OPT_K = 5
print("【test log】 OPT_K selected for downstream clustering:", OPT_K)

fig, ax = plt.subplots(figsize=(7, 4))
# elbow curve
ax.plot(list(K_RANGE), wcss_by_k, marker="o", color="steelblue", linewidth=2)
ax.set_xlabel("Number of Clusters (K)")
ax.set_ylabel("WCSS (Inertia)")
ax.set_title("K-Means elbow — TF-IDF Tags")
ax.grid(True, alpha=0.3)
plt.tight_layout()
plt.show()
```

WCSS falls from **~71,016** (**K = 2**) to **~59,205** (**K = 8**), with the incremental gain shrinking fastest between **K = 4 → 5 → 6** (e.g. **66,229 → 64,242 → 62,404**). **`OPT_K = 5`** balances granularity against over-fragmenting sparse tag space. ### 5.3.1 NHST — do tag clusters differ in catalogue activity?

§5.3 reported **`Mean is_active`** near **0.99** on **C0/C1/C2/C4** but **~0.10** on **C3**. This step tests whether cluster membership associates with **`is_active`** at catalogue scale.

**Hypothesis (α = 0.05):** H₀ — **`is_active`** is independent of **`Cluster_ID`**. H₁ — at least one cluster differs in activity rate.

**Methods (complementary):**

1. **χ²** test of independence on the **`Cluster_ID` × `is_active`** contingency table.
2. **Kruskal–Wallis** on binary **`is_active`** by cluster (rank-based check consistent with §2.6e).

Requires **`df_cluster`** from §5.3.



```python
# Section 5.3.1 — NHST: Cluster_ID vs is_active (chi-square + Kruskal–Wallis)
from scipy import stats

ALPHA = 0.05

if "df_cluster" not in globals():
    raise RuntimeError("Run §5.3 clustering cells first so df_cluster exists.")

# contingency table
active_tab = pd.crosstab(df_cluster["Cluster_ID"], df_cluster["is_active"])
active_tab.index.name = "Cluster_ID"
active_tab.columns = ["inactive", "active"]
print("【test log】 Cluster_ID x is_active counts:\n", active_tab.to_string())

chi2_stat, chi_p, chi_dof, chi_expected = stats.chi2_contingency(active_tab)
print(
    "【test log】 chi-square test of independence:",
    f"chi2={chi2_stat:.4f}, dof={chi_dof}, p={chi_p:.4e}, alpha={ALPHA}",
)

# Kruskal–Wallis on is_active (0/1) across clusters
cluster_groups = [
    df_cluster.loc[df_cluster["Cluster_ID"] == cid, "is_active"].values
    for cid in sorted(df_cluster["Cluster_ID"].unique())
]
kw_stat, kw_p = stats.kruskal(*cluster_groups)
print(
    "【test log】 Kruskal–Wallis is_active by Cluster_ID:",
    f"H={kw_stat:.4f}, p={kw_p:.4e}",
)

# per-cluster active rate table for prose
rate_tbl = (
    df_cluster.groupby("Cluster_ID")["is_active"]
    .agg(n="size", active_rate="mean")
    .reset_index()
)
print("\n【test log】 active_rate by Cluster_ID:\n", rate_tbl.to_string(index=False))
```

**Observation:**
- Both **χ²** and **Kruskal–Wallis** reject H₀ at α = 0.05 (**p ≪ 0.001** in the **test log**), so clusters are not exchangeable with respect to **`is_active`**.
- **C0, C1, C2, C4** show **`active_rate ≈ 0.99`** in the rate table, while **C3** is near **~0.10** on **~33k** rows — matching §5.3 visuals.

**Analysis:** TF-IDF tag clustering recovers statistically distinct engagement regimes, not only descriptive token themes. The cold **C3** mass explains why numeric-only models (Part 4) confuse paid shells with inactive listings.

**Next:** §5.4 interprets cluster personas and links back to §4.4 FP/FN patterns before Part 6 stress-tests priors on 2026 releases.

```python
# Section 5.3.1 — NHST: Cluster_ID vs is_active (chi-square + Kruskal–Wallis)
from scipy import stats

if "df_cluster" not in globals():
    raise NameError("Run §5.3 first — df_cluster with Cluster_ID is required.")

ALPHA = 0.05
cluster_nhst = df_cluster[["Cluster_ID", "is_active"]].dropna().copy()
cluster_nhst["Cluster_ID"] = cluster_nhst["Cluster_ID"].astype(int)
cluster_nhst["is_active"] = cluster_nhst["is_active"].astype(int)

ct = pd.crosstab(cluster_nhst["Cluster_ID"], cluster_nhst["is_active"])
print("【test log】 Cluster_ID x is_active counts:")
print(ct.to_string())

chi2, chi2_p, chi2_dof, _expected = stats.chi2_contingency(ct.values)
print(
    "【test log】 chi-square:",
    f"chi2={chi2:.4f}, df={int(chi2_dof)}, p={chi2_p:.4e}, alpha={ALPHA}",
)

kw_groups = [
    cluster_nhst.loc[cluster_nhst["Cluster_ID"] == cid, "is_active"].values
    for cid in sorted(cluster_nhst["Cluster_ID"].unique())
]
kw_stat, kw_p = stats.kruskal(*kw_groups)
print(
    "【test log】 Kruskal–Wallis on is_active by cluster:",
    f"H={kw_stat:.4f}, p={kw_p:.4e}, alpha={ALPHA}",
)

active_rate = cluster_nhst.groupby("Cluster_ID")["is_active"].mean().sort_index()
print("【test log】 observed active rate by cluster:")
print(active_rate.to_string())

```

**Observation**

- The **χ²** contingency table ( **test log** ) links **`Cluster_ID`** to **`is_active`**. With **five** clusters and **114,576** rows, **p ≪ 0.05** rejects H₀ — consistent with **C3 ~0.10** active vs **~0.99** on **C0/C1/C2/C4** in §5.3.
- **Kruskal–Wallis** on the same binary outcome yields the same conclusion ( **test log** ).

**Analysis**

Tag-based **K-Means** segments are statistically distinct on activity, not only on TF-IDF tokens. **C3** anchors the low-activity long tail, explaining numeric-only false positives in Part 4 and motivating cluster dummies in Part 6.

## Part 6: Hypothesis stress-test on the 2026 cohort

Chapters 4–5 established interpretable hypotheses: **commercial numerics** (§4) and **tag/theme clusters** (§5) relate differently to catalogue activity. This chapter does **not** build a production forecast. It **stress-tests** those hypotheses on listings released **on or after 2026-01-01** using strict time-forward validation.

**Design intent:** train on pre-2026 history, score on an **84-title external cohort**, and compare **Model A** (numeric-only) vs **Model B** (numeric + cluster dummies).

**Methodology note (label censoring):** On the 2026 slice, observed **`is_active = 0`** for every row because votes/CCU have not yet accumulated in this snapshot — **incubation**, not proof that every launch is permanently inactive. Under this **label censoring**, headline **accuracy** is a poor primary metric (predicting all inactive can yield **100%** accuracy trivially). Primary readouts are **(i) false-active count** (Model A vs B ablation) and **(ii) `p_active` diagnostic scores**, interpreted against historical cluster profiles — not as deployable launch forecasts.

### 6.1 Feature assembly and time-forward cohort split

**Purpose:** reuse the same prior numeric block and §5 cluster dummies from earlier chapters, then partition by **`Release date`** so no post-cutoff rows enter training.

**Numeric block (§4):** **`Price`**, **`Discount`**, **`DLC count`**, **`Achievements`**, **`Required age`**, **`Windows`**, **`Mac`**, **`Linux`**.

**Theme block (§5):** **`Cluster_ID`** → **`pd.get_dummies(..., prefix='cluster')`** (five **`cluster_*`** columns).

**Timeline (strict forward validation):** parse **`Release date`**; **history train** = releases **before 2026-01-01**; **2026 stress-test cohort** = **on/after 2026-01-01**. No random **`train_test_split`**.

Target **`y`** remains **`is_active`** from §4 (observed engagement proxy, not a deployment label).

```python
# Section 6.1 — multimodal X_fusion and time-based train / 2026 test split
from sklearn.ensemble import RandomForestClassifier
from sklearn.metrics import (
    accuracy_score,
    precision_score,
    recall_score,
    f1_score,
)

# prerequisite check
if "df_foundation" not in globals():
    raise NameError("df_foundation not found — run Part 3 through §3.4 first.")

NUMERIC_FEATURES = [
    "Price",
    "Discount",
    "DLC count",
    "Achievements",
    "Required age",
    "Windows",
    "Mac",
    "Linux",
]
RELEASE_CUTOFF = pd.Timestamp("2026-01-01")

# multimodal frame
df_multimodal = df_foundation.copy()
if "is_active" not in df_multimodal.columns:
    df_multimodal["is_active"] = (
        (df_multimodal["Positive"] + df_multimodal["Negative"] > 0)
        | (df_multimodal["Peak CCU"] > 0)
    ).astype(int)

# attach cluster labels
if "Cluster_ID" not in df_multimodal.columns:
    _cluster_src = globals().get("df_cluster")
    if (
        _cluster_src is not None
        and hasattr(_cluster_src, "columns")
        and "Cluster_ID" in _cluster_src.columns
    ):
        df_multimodal["Cluster_ID"] = _cluster_src["Cluster_ID"]
    else:
        raise KeyError(
            "Cluster_ID missing — run Part 5 (§5.1–5.3) before Part 6."
        )

# parse release date
df_multimodal["release_dt"] = pd.to_datetime(
    df_multimodal["Release date"], errors="coerce"
)
n_bad_dates = int(df_multimodal["release_dt"].isna().sum())
if n_bad_dates > 0:
    print("【test log】 rows dropped (unparsed Release date):", n_bad_dates)
    df_multimodal = df_multimodal.loc[df_multimodal["release_dt"].notna()].copy()

# one-hot clusters
cluster_dummies = pd.get_dummies(
    df_multimodal["Cluster_ID"], prefix="cluster", dtype=int
)
X_numeric = df_multimodal[NUMERIC_FEATURES].fillna(0)
# numeric + cluster dummies
X_fusion = pd.concat([X_numeric, cluster_dummies], axis=1)
y = df_multimodal["is_active"]

# pre-2026 train mask
is_history = df_multimodal["release_dt"] < RELEASE_CUTOFF
# 2026 external test mask
is_2026_test = df_multimodal["release_dt"] >= RELEASE_CUTOFF

X_num_hist = X_numeric.loc[is_history]
X_fusion_hist = X_fusion.loc[is_history]
y_hist = y.loc[is_history]

X_num_2026 = X_numeric.loc[is_2026_test]
X_fusion_2026 = X_fusion.loc[is_2026_test]
y_2026 = y.loc[is_2026_test]

print("【test log】 df_multimodal rows:", len(df_multimodal))
print("【test log】 X_numeric shape:", X_numeric.shape)
print("【test log】 cluster dummy columns:", list(cluster_dummies.columns))
print("【test log】 X_fusion shape:", X_fusion.shape)
print("【test log】 history train rows (< 2026-01-01):", len(X_num_hist))
print("【test log】 2026 external test rows (>= 2026-01-01):", len(X_num_2026))
print(
    "【test log】 history is_active rate:",
    round(float(y_hist.mean()), 4),
)
print(
    "【test log】 2026 observed is_active rate:",
    round(float(y_2026.mean()), 4),
)
```

**`X_fusion`** combines **8** numerics with **5** **`cluster_*`** dummies (**13** columns). The **history** window holds **114,492** rows (pre-2026); the **2026 stress-test cohort** has **84** listings. Observed **`is_active`** on 2026 titles is **0%** — **label censoring** from day-zero listings without vote/CCU history. §6.2–§6.3 therefore report **false-active ablation counts** and **`p_active` diagnostics**; they do **not** treat observed labels as a fair benchmark for predicting future engagement.

### 6.2 Ablation stress-test: numeric-only vs multimodal fusion

**Question under test:** does §4’s price/discount prior **over-generalise** on day-zero 2026 listings when §5 theme context is withheld?

Two **`RandomForestClassifier`** models share **`n_estimators=100`**, **`max_depth=6`**, **`min_samples_leaf=50`**, **`random_state=RANDOM_STATE`**: **Model A** (numeric only) vs **Model B** (**`X_fusion`**). Both train on **pre-2026 history** and score on the **2026 stress-test cohort**.

**Primary readout:** count of **false-active** hard predictions (`y_pred = 1` while observed **`is_active = 0`** under censoring) — Model A vs Model B.

**Secondary readout (reported but not headline):** Accuracy / Precision / Recall / F1 — included only to show that **active-class metrics are undefined** when no 2026 row is observed active; do **not** interpret Model B’s accuracy as model quality.

```python
# Section 6.2 — Random Forest ablation on 2026 stress-test cohort (false-active primary)
from IPython.display import HTML, display
from sklearn.ensemble import RandomForestClassifier
from sklearn.metrics import (
    accuracy_score,
    precision_score,
    recall_score,
    f1_score,
)

# kernel dependency check
_REQUIRED_62 = [
    "X_num_hist",
    "X_fusion_hist",
    "y_hist",
    "X_num_2026",
    "X_fusion_2026",
    "y_2026",
    "RANDOM_STATE",
]
_missing_62 = [name for name in _REQUIRED_62 if name not in globals()]
if _missing_62:
    raise NameError(
        "Run §6.1 first; missing in kernel: " + ", ".join(_missing_62)
    )

# shared RF hyperparams
RF_PARAMS = dict(
    n_estimators=100,
    max_depth=6,
    min_samples_leaf=50,
    random_state=RANDOM_STATE,
    n_jobs=-1,
)


# metric helper
def _classification_metrics(y_true, y_pred):
    return {
        "Accuracy": accuracy_score(y_true, y_pred),
        "Precision": precision_score(y_true, y_pred, pos_label=1, zero_division=0),
        "Recall": recall_score(y_true, y_pred, pos_label=1, zero_division=0),
        "F1-Score": f1_score(y_true, y_pred, pos_label=1, zero_division=0),
    }


# numeric-only model
model_a = RandomForestClassifier(**RF_PARAMS)
# fusion model
model_b = RandomForestClassifier(**RF_PARAMS)

# train A on history
model_a.fit(X_num_hist, y_hist)
model_b.fit(X_fusion_hist, y_hist)

# 2026 predictions
pred_a_2026 = model_a.predict(X_num_2026)
pred_b_2026 = model_b.predict(X_fusion_2026)

metrics_a = _classification_metrics(y_2026, pred_a_2026)
metrics_b = _classification_metrics(y_2026, pred_b_2026)

n_2026 = len(y_2026)
false_active_a = int((pred_a_2026 == 1).sum())
false_active_b = int((pred_b_2026 == 1).sum())

# primary ablation readout + secondary classification metrics
ablation_df = pd.DataFrame(
    [
        {
            "Model": "A — numeric only",
            "False_active_pred": false_active_a,
            "Cohort_n": n_2026,
            **metrics_a,
        },
        {
            "Model": "B — multimodal fusion",
            "False_active_pred": false_active_b,
            "Cohort_n": n_2026,
            **metrics_b,
        },
    ]
).round(4)

col_order = [
    "Model",
    "False_active_pred",
    "Cohort_n",
    "Accuracy",
    "Precision",
    "Recall",
    "F1-Score",
]
ablation_df = ablation_df[col_order]

print("【test log】 ablation primary — false-active hard predictions (label-censored cohort):")
print(f"  Model A: {false_active_a} / {n_2026}")
print(f"  Model B: {false_active_b} / {n_2026}")
print("【test log】 secondary metrics (accuracy misleading when all labels are inactive):")
print(ablation_df.to_string(index=False))

ablation_html = ablation_df.to_html(index=False, float_format="%.4f")
display(HTML(ablation_html))

# save for §6.3
rf_fusion_2026 = model_b
```
